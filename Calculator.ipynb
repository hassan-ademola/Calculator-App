```python
from tkinter import Tk,Frame,Button,Entry,END
window = Tk()
window.title('Calculator')
window.geometry('250x350')

canvas = Frame(window, bg = 'white', bd = 0)
canvas.place(relwidth = 1, relheight = 1)

def zero():
    entry.insert(END, '0')
def one():
    entry.insert(END, '1')
def two():
    entry.insert(END, '2')
def three():
    entry.insert(END, '3')
def four():
    entry.insert(END, '4')
def five():
    entry.insert(END, '5')
def six():
    entry.insert(END, '6')
def seven():
    entry.insert(END, '7')
def eight():
    entry.insert(END, '8')
def nine():
    entry.insert(END, '9')
def plus():
    entry.insert(END, '+')
def point():
    entry.insert(END, '.')
def minus():
    entry.insert(END, '-')
def multiply():
    entry.insert(END, '*')
def divide():
    entry.insert(END, '/')
def clear():
    entry.delete(0, END)
    entry.config(font = ('TkDefaultFont', 16), fg ='black')
def quit():
    window.destroy()

def equals():
    result = round(eval(entry.get()),7)
    entry.delete(0, END)
    entry.config(font = ('TkDefaultFont', 35), fg ='#ffc23b')
    entry.insert(END, result)


#row1
zero = Button(canvas, text = '0', command = zero, bd = 0, font = ('TkDefaultFont', 12))
zero.place(relx = 0.04, rely = 0.85, relwidth = 0.44, relheight = 0.12)
point = Button(canvas, text = '.', command = point, bd = 0, font = ('TkDefaultFont', 15, 'bold'))
point.place(relx = 0.52, rely = 0.85, relwidth = 0.2, relheight = 0.12)
equals = Button(canvas, text = '=', command = equals, bd = 0, font = ('TkDefaultFont', 15), bg = '#ffc23b', fg = '#cc9423')
equals.place(relx = 0.76, rely = 0.7, relwidth = 0.2, relheight = 0.27)

#row2
one = Button(canvas, text = '1', command = one, bd = 0, font = ('TkDefaultFont', 12))
one.place(relx = 0.04, rely = 0.7, relwidth = 0.2, relheight = 0.12)
two = Button(canvas, text = '2', command = two, bd = 0, font = ('TkDefaultFont', 12))
two.place(relx = 0.28, rely = 0.7, relwidth = 0.2, relheight = 0.12)
three = Button(canvas, text = '3', command = three, bd = 0, font = ('TkDefaultFont', 12))
three.place(relx = 0.52, rely = 0.7, relwidth = 0.2, relheight = 0.12)

#row3
four = Button(canvas, text = '4', command = four, bd = 0, font = ('TkDefaultFont', 12))
four.place(relx = 0.04, rely = 0.55, relwidth = 0.2, relheight = 0.12)
five = Button(canvas, text = '5', command = five, bd = 0, font = ('TkDefaultFont', 12))
five.place(relx = 0.28, rely = 0.55, relwidth = 0.2, relheight = 0.12)
six = Button(canvas, text = '6', command = six, bd = 0, font = ('TkDefaultFont', 12))
six.place(relx = 0.52, rely = 0.55, relwidth = 0.2, relheight = 0.12)
minus = Button(canvas, text = '-', command = minus, bd = 0, font = ('TkDefaultFont', 15), bg = '#ffeec2', fg = '#ca9225')
minus.place(relx = 0.76, rely = 0.55, relwidth = 0.2, relheight = 0.12)

#row4
seven = Button(canvas, text = '7', command = seven, bd = 0, font = ('TkDefaultFont', 12))
seven.place(relx = 0.04, rely = 0.4, relwidth = 0.2, relheight = 0.12)
eight = Button(canvas, text = '8', command = eight, bd = 0, font = ('TkDefaultFont', 12))
eight.place(relx = 0.28, rely = 0.4, relwidth = 0.2, relheight = 0.12)
nine = Button(canvas, text = '9', command = nine, bd = 0, font = ('TkDefaultFont', 12))
nine.place(relx = 0.52, rely = 0.4, relwidth = 0.2, relheight = 0.12)
plus = Button(canvas, text = '+', command = plus, bd = 0, font = ('TkDefaultFont', 15), bg = '#ffeec2', fg = '#ca9225')
plus.place(relx = 0.76, rely = 0.4, relwidth = 0.2, relheight = 0.12)

#sympols
quit = Button(canvas, text = 'Quit', command = quit, bd = 0, font = ('TkDefaultFont', 12), bg = '#ffeec2', fg = '#ca9225')
quit.place(relx = 0.04, rely = 0.25, relwidth = 0.2, relheight = 0.12)
clear = Button(canvas, text = 'C', command = clear, bd = 0, font = ('TkDefaultFont', 12), bg = '#ffeec2', fg = '#ca9225')
clear.place(relx = 0.28, rely = 0.25, relwidth = 0.2, relheight = 0.12)
divide = Button(canvas, text = '/', command = divide, bd = 0, font = ('TkDefaultFont', 14), bg = '#ffeec2', fg = '#ca9225')
divide.place(relx = 0.52, rely = 0.25, relwidth = 0.2, relheight = 0.12)
multiply = Button(canvas, text = 'x', command = multiply, bd = 0, font = ('TkDefaultFont', 14), bg = '#ffeec2', fg = '#ca9225')
multiply.place(relx = 0.76, rely = 0.25, relwidth = 0.2, relheight = 0.12)

entry = Entry(canvas, font = ('TkDefaultFont', 14), bd = 0, justify = 'right')
entry.place(relx = 0.04, rely = 0.03, relheight = 0.2, relwidth = 0.92)
entry.focus_set()
window.mainloop()
```
